# Exploratory Data Analysis(EDA)

```python
# since it would be cluttered to plot sales against date for all store so I will pick a random store for the sake of visualisations

store_1111_data = train_data[train_data["Store"] == 1111]
fig = px.scatter(store_1111_data, x="Date" , y="Sales",
             hover_data={"Date": "|%B %d,%Y"},color="Open",
             title='Daily Sales for Store 1111(store was open/closed)')
fig.update_xaxes(dtick="M1",tickformat="%b\n%Y")
fig.update_traces(marker=dict(color=['green' if i==1 else 'red' for i in store_1111_data['Open']]))
fig.show()
```
